# MATRIX CHAIN MULTIPLICATION Dynamic Programming with table filling.
Problem Description:
Find minimum number of scalar multiplications needed to multiply a chain of matrices..

#Theory
Cost of multiplying two matrices:
If A is p×q and B is q×r, cost = p × q × r

Dynamic Programming idea:
Break problem into smaller subproblems.
Store results in a table to avoid recomputation.

m[i][j] = minimum cost to multiply Ai to Aj

## Algorithm
1. Take array p[] of size n
2. Create table m[n][n]
3. Set m[i][i] = 0
4. Fill table diagonally
5. For each chain length L:
    Compute m[i][j]
    Try all possible k between i and j
6. Choose minimum cost
7. Return m[1][n-1]

```python
import sys

def matrix_chain_multiplication(p):
    n = len(p)

    # Create DP table
    m = [[0 for _ in range(n)] for _ in range(n)]

    # L is chain length
    for L in range(2, n):
        for i in range(1, n - L + 1):
            j = i + L - 1
            m[i][j] = sys.maxsize

            for k in range(i, j):
                cost = (m[i][k] +
                        m[k+1][j] +
                        p[i-1] * p[k] * p[j])

                if cost < m[i][j]:
                    m[i][j] = cost

    return m

```

```python
# Input dimensions
p = [10, 30, 5, 60]

table = matrix_chain_multiplication(p)

print("Minimum multiplication cost:", table[1][len(p)-1])

print("\nDP Table:")
for row in table[1:]:
    print(row[1:])
```

# Exercise
1. Given p = [5, 10, 3, 12, 5, 50, 6]
2. Compute minimum cost
3. Print optimal parenthesization
4. Compare number of multiplications saved

```python
#Q1

p = [5, 10, 3, 12, 5, 50, 6]
import sys

def matrix_chain_multiplication(p):
    n = len(p)

    m = [[0]*n for _ in range(n)]
    s = [[0]*n for _ in range(n)]

    for L in range(2, n):
        for i in range(1, n-L+1):
            j = i + L - 1
            m[i][j] = sys.maxsize

            for k in range(i, j):
                cost = m[i][k] + m[k+1][j] + p[i-1]*p[k]*p[j]

                if cost < m[i][j]:
                    m[i][j] = cost
                    s[i][j] = k

    return m, s
```

```python
#Q2

def print_optimal(s, i, j):
    if i == j:
        print(f"A{i}", end="")
    else:
        print("(", end="")
        print_optimal(s, i, s[i][j])
        print_optimal(s, s[i][j] + 1, j)
        print(")", end="")
```

```python

#Q3

p = [10, 30, 5, 60]

m, s = matrix_chain_multiplication(p)

print("Minimum multiplication cost:", m[1][len(p)-1])

print("\nOptimal Parenthesization:", end=" ")
print_optimal(s, 1, len(p)-1)

print("\n\nDP Table:")
for row in m[1:]:
    print(row[1:])
```

```python
#Q4
p = [5, 10, 3, 12, 5, 50, 6]

m, s = matrix_chain_multiplication(p)

print("Minimum multiplication cost:", m[1][len(p)-1])

print("Optimal Parenthesization:", end=" ")
print_optimal(s, 1, len(p)-1)
print()
```
